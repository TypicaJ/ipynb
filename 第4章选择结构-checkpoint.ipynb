## 4.2　布尔表达式

每条if语句的核心都是一个值为True或False的表达式，这种表达式被称为布尔表达式。 Python根据条件测试的值为True还是False来决定是否执行if语句中的代码。如果布尔表达式的值为True，Python就执行紧跟在if语句后面的代码；如果为False， Python就忽略这些代码。

### 4.2.1　条件语句示例


```python
a = 5
b = 6
if (a > b):
    print(a + b)
elif(a==b):
    print(a**2)
else:
    print(a * b)
```

* 真假判断：0，None为假，空对象为假，包括''，()，[]，{}

```python
if '':
    print('真')
else:
    print('假')
```

```python
if ' ':
    print('真')
else:
    print('假')
```

###  4.2.2　等于判断 ==

大多数条件测试都将一个变量的当前值同特定值进行比较。最简单的条件测试检查变量的值是否与特定值相等

```python
a = 5
b = 6
if (a == b):
    print(a + b)
else:
    print(a * b)
```

* 注意注意:一个等号是赋值!!! 两个等号才是比较运算符，用来判断是否相等。  
比如b=6的意思是，将6这个整数值赋给变量b,而不是判断相等。

### 4.2.3　不等于判断

要判断两个值是否不等，可结合使用惊叹号和等号（ !=），其中的惊叹号表示不，在很多编程语言中都如此。

```python
a = 5
b = 6
if (a != b):
    print(a + b)
else:
    print(a * b)
```

条件表达式是a!=b，意思是判断变量a的值和变量b的值是否不相等。如果不相等则表达式的值为True，否则为False。

## 4.2　单分支选择结构

### 4.2.1　单分支语句

最简单的if语句只有一个测试和一个操作，语法如下：

```python
if (conditional_expression):
     do something
```

要判断一个变量a的值是否小于0，如果是，则将其改为正值，代码如下：


```python
a = -5
if (a < 0):
    a = -1 * a #还可以用什么？
    print(a)

    

print(a)
```

在if语句中，缩进的作用和其他语言中的括号类似。如果测试通过了，将执行if语句后面所有缩进的代码行，否则将忽略它们。

## 4.3　双分支选择结构

## 4.3.1　双分支语句

单分支语句只是在条件表达式满足的时候执行一些操作，条件不满足的时候直接跳过这些操作而执行下面的代码。但经常需要在条件测试通过时执行一个操作，并在没有通过时执行另一个操作；在这种情况下，可使用Python提供的if-else语句。

if-else语句块类似于简单的if语句，但其中的else语句让你能够指定条件测试未通过时要执行的操作，我们称这种结构为双分支语句。

判断一个变量是奇数还是偶数，代码如下：

```python
a = 5
if (a % 2 == 1):
    print('奇数')
else:
    print('偶数')
```

希望为受某些条件限制的变量赋值。例如，如果x大于0，则以下语句将1赋给y；如果x小于0或等于0，则将-1赋给y。如下例：

```python
x = -4
if (x > 0):
    y = 1 
else:
    y = -1
print(y)
```

这种双分支赋值语句有一种简化的写法，叫做条件表达式。条件表达式的格式是：

变量 = 表达式1 if 布尔表达式 else 表达式2

```python
x = -4
y = 1 if x > 0 else -1
print(y)
#讲到这儿了
```

这种表达式的优势是程序变得更加简洁。也可以把刚才的奇偶判断的例子写成：

```python
a = 5
print('奇数') if (a % 2 == 1) else print('偶数')
```

## 4.4　多分支选择结构

### 4.4.1　多分支语句

超过两个的情形，可使用Python提供的if-elif-else结构。 

Python只执行if-elif-else结构中的一个代码块，它依次检查每个条件测试，直到遇到通过了的条件测试。测试通过后， Python将执行紧跟在它后面的代码，并跳过余下的测试。

下面程序用来判断一个数是正数，负数还是0。

```python
a = 5
if (a > 0):
    print('正数')
elif (a < 0):
    print('负数')
else:
    print('零')
```

判断一个学生的成绩是属于什么档次，0到59为不及格，60-74为中等，75-89为良好，90-100为优秀，其他为非法情况。


```python
score = 88
if (score > 100 or score < 0):
    print('异常成绩')
elif (score >= 90):
    print('优秀')
elif (score >= 75):
    print('良好')
elif (score >= 60):
    print('中等')
else:
    print('不及格')
```

Python并不要求if-elif结构后面必须有else代码块。在有些情况下， else代码块很有用；而在其他一些情况下，使用一条elif语句来处理特定的情形更清晰：

```python
score = 88
if (score > 100 or score < 0):
    print('异常成绩')
elif (score >= 90):
    print('优秀')
elif (score >= 75):
    print('良好')
elif (score >= 60):
    print('中等')
elif (score >= 0):
    print('不及格')
```

else是一条包罗万象的语句，只要不满足任何if或elif中的条件测试，其中的代码就会执行，这可能会引入无效甚至恶意的数据。如果知道最终要测试的条件，应考虑使用一个elif代码块来代替else代码块。这样，就可以肯定，仅当满足相应的条件时，代码才会执行。

## 4.5　选择结构的嵌套

### 4.5.1　分支语句的嵌套 

一个if语句可以放置在另一个if语句中，以形成嵌套的if语句。if或if else语句中的语句可以是任何合法的Python语句，包括另一个if或if else语句。内部if语句被称为嵌套在外部if语句。内部if语句可以包含另一个if语句；实际上，嵌套深度没有任何限制。

```python
a = 5
if (a > 0):    
    print('正数')
else:
    if (a < 0):
        print('负数')
    else:
        print('零')
```

分支语句的嵌套通常用来实现一些多分支的语句。接下来，我们继续把上节中输出成绩的例子，改为嵌套形式，如下所示：

```python
score = 88
if (score > 100 or score < 0):
    print('异常成绩')
else:
    if (score >= 90):
        print('优秀')
    else:
        if (score >= 75):
            print('良好')
        else:
            if (score >= 60):
                print('中等')
            else:
                print('不及格')
```

使用嵌套的方式会有两个缺点：

* 1.可读性差；
* 2.容易出错；  
所以如果不是必要的情况下，尽量不要使用嵌套。  
灵活运用elif可以减少嵌套。

### 4.5.2　分支语句常见错误

1.缩进问题

```python
a = -5
if (a < 0):
    a = -1 * a
print(a)
```

但如果不小心写成：

```python
a = -5
if (a < 0):
    a = -1 * a
    print(a)
```

这种情况下，只有a小于0的时候才会输出，a大于0的时候不会输出a的值。

2.边界问题

在写布尔表达式来判断范围时，经常在边界是否等于的问题上犯错，比如我们上节中的分数判断例子，如果写成如下的情况，就会有问题：


```python
score = 60
if (score > 100 or score < 0):
    print('异常成绩')
elif (score > 90):
    print('优秀')
elif (score > 75):
    print('良好')
elif (score > 60):
    print('中等')
else:
    print('不及格')
```

还有一个不算错误的初学者常见问题就是冗余判断，如下例：

```python
score = 88
if (score > 100 or score < 0):
    print('异常成绩')
elif (score >= 90 and score <= 100):
    print('优秀')
elif (score >= 75 and score < 90):
    print('良好')
elif (score >= 60 and score < 75):
    print('中等')
elif (score >= 0 and score < 60):
    print('不及格')
```

3.条件遗漏问题

当需要多个嵌套的分支语句来判断复杂问题时，容易因为缺失条件，导致出错。接下来学习经典的判断是否闰年问题。

#判断一个年份是否是闰年；能被4整除且不能被100整除或者能被400整除

#能被4整除、但不能被100整除的年份都是闰年，比如2008/2012/2048年，或能被400整除的年份为闰年，2000年不符合这两个条件的年份不是闰年


闰年的定义是：如果是世纪年，如1900，2000等可以被100整除的年份，只有可以被400整除的年份是闰年，其他年份是平年；其他不是世纪年的年份可以被4整除的就是闰年，否则就是平年。 正确的写法比如：


```python
year = 2018
if (year % 4) == 0:
    if (year % 100) == 0:
        if (year % 400) == 0:
            print("是闰年")   # 整百年能被400整除的是闰年
        else:
            print("不是闰年")
    else:
        print("是闰年")       # 非整百年能被4整除的为闰年
else:
    print("不是闰年")
```

但如果有条件遗漏，比如下例：

```python
year = 2100
if (year % 4) == 0:
    if (year % 100) == 0:
        if (year % 400) == 0:
            print("是闰年")   # 整百年能被400整除的是闰年
    else:
        print("是闰年")       # 非整百年能被4整除的为闰年
else:
    print("不是闰年")
```

在上例中，因为年份为2100年，不满足上面的各种分支情况，所以不会有任何输出。实际上，在这种情况下，为避免有条件遗漏，可以考虑把判断条件合成一条语句，则条件遗漏的可能性将会大大降低，如下例所示：

```python
year = 2100
if (year % 400 == 0 or (year % 4 == 0 and year % 100 != 0)):
    print("是闰年")      
else:
    print("不是闰年")

```
